### всё в одном дф

```python
# собираем всё в один дф
df = pd.concat([df_balashiha, df_chehov, df_dmitrov, df_dolgopa, 
                df_dubna, df_electrostal, df_istra, df_khimki, 
                df_krasnogorsk, df_lobnya, df_luberts, df_mitishi, df_moscow, df_podolsk,
                df_solnechnogorsk, df_zelenograd, df_pushkino, df_zaraizk, 
                df_mozhaisk, df_shelcovo, df_serpuhov, df_zhukovskiy, df_odintsovo])
```

```python
# инфа до удаления дубликатов
df.info()
```

```python
# дроп дубликатов
df = df.drop_duplicates()
```

```python
# инфа после дропа дубликатов
df.info()
```

```python
# все города из дф
print(df['location'].unique())
```

```python
df['location'] = df['location'].replace({'Серпухов}': 'Серпухов', 'Химки}': 'Химки'})
```
